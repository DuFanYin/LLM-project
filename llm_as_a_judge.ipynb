```python
bm_df = bm_df.sample(frac=1).reset_index(drop=True)
# ft_df = ft_df.sample(frac=1).reset_index(drop=True)
rl_df = rl_df.sample(frac=1).reset_index(drop=True)
```

```python
bm_df.to_json("bm_llm.json", orient="records", indent=4)

# ft_df.to_json("ft_llm.json", orient="records", indent=4)

rl_df.to_json("rl_llm.json", orient="records", indent=4)
```
